```python
# Create TF frozen graph and Save pb
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.python.framework.convert_to_constants import convert_variables_to_constants_v2
import os 

model_name = "efficientnetb7"
COLAB_PATH = "/content/drive/MyDrive/Colab Notebooks/大伯 - ML_Code/"
h5_path = COLAB_PATH + "efficientnetb7_model.h5"

model = tf.keras.models.load_model(h5_path) 

# get model TF graph
model_graph = tf.function(lambda x: model(x))

# get concrete function
model_graph = model_graph.get_concrete_function(
    tf.TensorSpec(model.inputs[0].shape, model.inputs[0].dtype))

# obtain frozen concrete function
frozen_func = convert_variables_to_constants_v2(model_graph)
# get frozen graph
frozen_func.graph.as_graph_def()

# save full tf model
tf.io.write_graph(graph_or_graph_def=frozen_func.graph,
                  logdir=COLAB_PATH,
                  name=f"{model_name}.pb",
                  as_text=False)
print("save to pb !")

tf.io.write_graph(graph_or_graph_def=frozen_func.graph,
                  logdir=COLAB_PATH, 
                  name=f"{model_name}.pbtxt", 
                  as_text=True)
print("save to pbtxt !")
```
